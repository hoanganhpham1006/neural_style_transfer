```python
from matplotlib import pyplot as plt
import cv2
```

```python
def read_img(img_path):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
#Styles

img1 = read_img('style_1.jpg')
img2 = read_img('style_2.jpeg')
img3 = read_img('style_3.jpg')
```

```python
plt.figure(figsize=(16, 16))

plt.subplot(3, 3, 1)
plt.imshow(img1)

plt.subplot(3, 3, 2)
plt.imshow(img2)

plt.subplot(3, 3, 3)
plt.imshow(img3)

plt.show()
```

```python
#Origin images

plt.figure(figsize=(16, 16))

for i in range(3):
    img = read_img('origin_' + str(i+1) + '.jpg' )
    plt.subplot(3, 3, i+1)
    plt.imshow(img)

plt.show()
```

```python
#Image 1

list_img = ['origin_1/27.png',
           'origin_1/28.png',
           'origin_1/29.png',
           'origin_1/30.png',
           'origin_1/31.png',
           'origin_1/32.png',
           'origin_1/33.png',
           'origin_1/35.png',
           'origin_1/36.png']
```

```python
fig=plt.figure(figsize=(12, 12))
columns = 3
rows = 3
for i in range(1, columns*rows +1):
    img = read_img(list_img[i-1])
    fig.add_subplot(rows, columns, i)
    plt.title("Iteration " + str(i))
    plt.imshow(img)
plt.show()
```
